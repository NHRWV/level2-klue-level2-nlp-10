```python
from koeda import EDA
```

```python
eda = EDA(
    morpheme_analyzer="Okt", alpha_sr=0.3, alpha_ri=0.3, alpha_rs=0.3, prob_rd=0.3
)

text = "아버지가 방에 들어가신다"

result = eda(text)
print(result)
# eda 결과 매번 다름
# 아버지가 정실에 들어가신다 

result = eda(text, p=(0.9, 0.9, 0.9, 0.9), repetition=2)
print(result)
# ['아버지가 객실 아빠 안방 방에 정실 들어가신다', '아버지가 탈의실 방 휴게실 에 안방 탈의실 들어가신다']
```

```python
dataset_dir = "/opt/ml/dataset/train/train.csv"
pd_dataset = pd.read_csv(dataset_dir)
```

```python
pd_dataset
```

```python
# 모든 행을 돌면서, eda(sentence)
# eda된 문장에 subject/object 단어가 없거나 원래 문장과 동일하면, 다시 eda
for i in range(pd_dataset.shape[0]):
    r = pd_dataset.iloc[i]
    sen = eda(r["sentence"])
    while (eval(r['subject_entity'])['word'] not in sen) or (eval(r['object_entity'])['word'] not in sen) or (r["sentence"]==sen):
        sen = eda(r["sentence"])
    
    pd_dataset["sentence"].iloc[i] = sen
```

```python
# 중복 문장 및 mislabeled 문장 제거
idx = pd_dataset[(pd_dataset['id'].isin([858, 7080, 20838, 6352,25094, 10320, 277, 22258, 8364, 6749, 18458]))].index
pd_dataset = pd_dataset.drop(idx)
```
